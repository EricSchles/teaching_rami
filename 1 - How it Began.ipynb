# An introduction to data science

Data science is the culmination of a grand synthesis.  It is the recognition that all sciences are related by the study of data.  It is beyond statistics, a language of science.  It is in some ways seperate from mathematics because it is deeply applied, but in many ways it has become the new language of science and through it programming has become universalized.

Data science is the science of information.

But what is information?  The wonderful thing about it, is it has come about in many, many ways.

## Major Concepts

* Information Theory
* Probability Theory

* Linear Algebra

* The Calculus

* Software Engineering

As you can see from the above list, a Scientist of Data must be versed in many things.

## Some Algorithms 

* _K-Means clustering_ - From Computer Science and Information Theory

They used K-means to bucket the data into different segments.  Without this they'd have to deal with everything at once and it would be impossible to discern any pattern, because there is simply too much data.

* _Principal Component Analysis_ - From Linear Algebra

They used Principal Component Analysis (PCA) to be able to join and combine columns of data together, to improve the performance of their models, allowing for great precision of results.  This allowed for clearly theory, with higher levels of justification.

* _Generalized Linear Regression_ - from Statistics

This generalized linear model was highly interpretability and highly flexible so it could handle a number of cases that ordinary least squares could not.  

* _Support Vector Machines_ - from Linear Algebra

This model allowed scientists to, for the first time model data in a way beyond the constraints of linearity and normality assumed by linear models.

* _Decision Trees and it's variants_ - from Information Theory

Decision Trees are so called universal approximators.  They require no assumption about the underlying data.  In short, they can model anything where there is a relationship, regardless of the underlying topology of the data.  The data need not be normally distributed.  It need not even come from a euclidean space.  

* _Genetic Algorithms_ - Algebra, Computer Science, biology

Genetic algorithms are another class of universal approximators that simulate a process by decomposing it into component operations.  Using a genetic algorithm, you can create a function that will fit literally anything that can be fit by a mathematical function.

The final form of a genetic algorithm doesn't always make sense, but there is some interpretability here.

* _Neural Networks_ - From Artificial Intelligence and (Originally Control Theory)

Neural networks are universal approximators as well, but they don't search as much of the space as genetic algorithms do.  Typically initialization matters a lot here.  However, when trained correctly, neural networks are really great for modeling any kind of data.
